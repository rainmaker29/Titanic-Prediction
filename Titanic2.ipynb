```python
# Decreasing the titles by replacing with some common titles for the related titles
```

```python
for dataset in combine:
    dataset.Title = dataset.Title.replace(['Lady', 'Capt', 'Col',
    'Don', 'Dr', 'Major', 'Rev', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace(['Countess', 'Lady', 'Sir'], 'Royal')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
```

```python
#map each of the title groups to a numerical value
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Royal": 5, "Rare": 6}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
target = train['Survived']
```

```python
train.drop(['PassengerId','Survived'],inplace=True,axis=1)
```

```python
train.head(1)
```

```python
x_train,x_val,y_train,y_val = train_test_split(train,target,test_size=0.22,random_state=0)
```

```python
# Implementing Kneighbors
```

```python
y_train.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
pred=neigh.fit(x_train,y_train).predict(x_val)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_val,pred,normalize=True))
```

```python
from sklearn.svm import SVC

svc = SVC()
pred = svc.fit(x_train, y_train).predict(x_val)
print(accuracy_score(y_val,pred,normalize=True))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbk = GradientBoostingClassifier()
```

```python
pred = gbk.fit(x_train,y_train).predict(x_val)
print(accuracy_score(y_val,pred,normalize=True))
```
